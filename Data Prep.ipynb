```python
import pandas as pd
import numpy as np
import glob
import os
```

```python
os.getcwd()
```

```python
table_paths = glob.glob('C:\\Users\\tobij\\Desktop\\Ipynb Tasks\\Giredestrant/*.tab')
table_paths
```

```python
dfs = []
for item in table_paths:
    dfs.append(pd.read_csv(item, delimiter='\t'))
```

```python
df = pd.concat(dfs, axis=1)
```

```python
df.head()
```

```python
df.columns
```

```python
np.where(df.isna().sum())
```

```python
counts = df['count']
```

```python
counts_cols = counts.columns
```

```python
new_names = []
for idx, item in enumerate(counts_cols):
    if idx < 4:
        new_names.append(f'dmso_control_{idx + 1}')
    elif idx >= 4 and idx < 8:
        new_names.append(f'dmso_GPX4_{idx - 3}')
    elif idx >= 8 and idx < 12:
        new_names.append(f'giredestrant_control_{idx - 7}')
    else:
        new_names.append(f'giredestrant_GPX4_{idx - 11}')
```

```python
new_names
```

```python
counts.columns = new_names
```

```python
counts
```

```python
genes = df.iloc[:, 0]
```

```python
genes = genes.to_frame()
genes.head()
```

```python
df = pd.concat([genes, counts], axis=1)
```

```python
df.head()
```

```python
df.shape
```

```python
df.to_csv('counts.csv', index=False)
```

```python
df.columns
```

```python
controls = df[['name', 'dmso_control_1', 'dmso_control_2', 'dmso_control_3', 'dmso_control_4', 'giredestrant_control_1', 'giredestrant_control_2', 'giredestrant_control_3', 'giredestrant_control_4']]
```

```python
controls_cols = controls.columns
```

```python
gpx4s = df.loc[:, ~df.columns.isin(controls_cols)]
```

```python
gpx4s = pd.concat([genes, gpx4s], axis=1)
```

```python
controls.to_csv('controls.csv', index=False)
gpx4s.to_csv('gpx4s.csv', index=False)
```
